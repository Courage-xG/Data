```python
#import modules
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
from matplotlib.patches import Arc, Rectangle, Circle
```

```python
df = pd.read_csv('C:\\Users\\HP\\Downloads\\MYDATASPORT (4).csv') 
df
```

```python
#plotting Kelechi Onwe's progressive pass data

fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

# Define the pitch type
pitch_type = 'statsbomb'  # or 'opta'

# Create the pitch based on the type
pitch = Pitch(
    pitch_type=pitch_type,
    pitch_color='#22312b',
    line_color='#c7d5cc',
)

# Draw the pitch on the ax figure as well as invert the axis for this specific pitch
pitch.draw(ax=ax)

# Conditionally invert the y-axis based on pitch type
if pitch_type == 'statsbomb':
    plt.gca().invert_yaxis()  # Only invert for statsbomb pitch type

# Adjust coordinates based on pitch type
x_scale = 1.2 if pitch_type == 'statsbomb' else 1
y_scale = 0.8 if pitch_type == 'statsbomb' else 1
    
# Plot each pass with the corrected condition
for i in range(len(df['X player from'])):
    x_start = df['X player from'].iloc[i] * x_scale
    y_start = df['Y player from'].iloc[i] * y_scale
    x_end = df['X player to'].iloc[i] * x_scale
    y_end = df['Y player from'].iloc[i] * y_scale
    
    if df['Action'].iloc[i] == 'Successful Progressive Pass' and df['Player from'].iloc[i] == 'Kelechi Onwe':
        plt.plot(('X player from', 'X player to'), ('Y player from', 'Y player to'), color='green')
        plt.scatter('X player from', 'Y player from', color='green')
    elif df['Action'].iloc[i] == 'Unsuccessful Progressive Pass' and df['Player from'].iloc[i] == 'Kelechi Onwe':
        plt.plot(('X player from', 'X player to'), ('Y player from', 'Y player to'), color='red')
        plt.scatter('X player from', 'Y player from', color='red')

plt.title('Progressive Passes by Kelechi Onwe Successful Progressive Pass: Green | Unsuccessful Progressive Pass: Red',
    color='white', size=13)
plt.show()
```
